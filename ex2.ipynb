## Aufgabe 2
- Initialisierung der symmetrischen $20\times 20$ Wechselwirkungsmatrix $J_{ij}$
- Erzeugung durch `np.random.normal()` mit $$p(x)=\frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$
- WÃ¤hle $\mu = -3$ und $\sigma = \frac{1}{\sqrt{2}}$

```python
d = Counter()
for i in range(10000):
    J = network.random_interaction(20, -3, 1/np.sqrt(2))
    eig = np.real(np.linalg.eigvals(J + 0j))
    eig2 = np.round(eig, 1)
    unique, counts = np.unique(eig2, return_counts=True)
    d.update(dict(zip(unique, counts)))
```

```python
fig, ax = plt.subplots(figsize=(16/2.54, 8/2.54))
ax.set_xlabel('Eigenwerte')
ax.set_ylabel('Anzahl')
ax.plot(d.keys(), d.values(), '.')
ax.set_xlim(-10, 10)
ax.grid(True)

fig, ax = plt.subplots(figsize=(16/2.54, 7/2.54))
ax.set_xlabel('Eigenwerte')
ax.set_ylabel('Anzahl')
ax.plot(d.keys(), d.values(), '.')
ax.set_xlim(-70, -50)
ax.set_ylim(-10, 500)
x = np.arange(-65, -55, 0.1)
y = 400 * np.exp(-((x + 60.15)**2) / 2.0)
ax.plot(x, y, '--', linewidth=1.5)
ax.grid(True)
```
