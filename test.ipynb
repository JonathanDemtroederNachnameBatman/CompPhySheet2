```python
import numpy as np
%load_ext autoreload
%autoreload 2

import network
from ipywidgets import IntProgress
from IPython.display import display
```

```python
p = network.create_protein(interaction_type='normal')
network.plot_protein(p)
```

```python
folds = 5000
f = IntProgress(min=0, max=100)
display(f)
count = 0
one = folds // 100
while count <= folds:
    if count % one == 0:
        f.value += 1 
    p.random_fold_step(5)
    count += 1
print(f'Folds: {p.actual_folds}')
network.plot_protein(p)
```
